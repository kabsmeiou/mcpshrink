```python
import nlpaug.augmenter.word.back_translation as naw

def backtranslate (query: str) -> str:
    aug = naw.BackTranslationAug(
        from_model_name="Helsinki-NLP/opus-mt-ko-en",
        to_model_name="Helsinki-NLP/opus-mt-en-ko",
    )

    augmented_query = aug.augment(query)
    return augmented_query
```

```python
backtranslate("Hello. My name is Kane. It's nice to meet you.")
```
